```python
import sys
sys.path.append("./scripts")

from functions_basic import *

raw_data = pd.read_csv("data/gene_game/progressive_death_ratio_0_2,5.csv")
raw_data
```

```python
# have a look at this data through histograms and other graphs
# see how you can start going through the book and think of interesting ideas to investigate
data = [raw_data.iloc[201*j:201*(j+1), :].copy() for j in range(6)]
data[0]
```

```python
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.plot(data[i].time, data[i].animal_number, label=f"Death factor = {0.5*i}")
    
plt.xlabel("Time (turns)")
plt.ylabel("Population number")
plt.title("Population evolution for different death factors")
plt.legend()
plt.show()
```

```python
# for genes:
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.plot(data[i].time, data[i].gene_1_mean, label=f"Death factor = {0.5*i}")
    
plt.xlabel("Time (turns)")
plt.ylabel("Mean of gene_1")
plt.title("Mean of gene_1 for different death factors")
plt.legend()
plt.show()
```

```python
# for genes:
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.plot(data[i].time, data[i].gene_1_std, label=f"Death factor = {0.5*i}")
    
plt.xlabel("Time (turns)")
plt.ylabel("Std of gene_1")
plt.title("Std of gene_1 for different death factors")
plt.legend()
plt.show()
```

```python
# for genes:
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.plot(data[i].time, data[i].gene_0_std, label=f"Death factor = {0.5*i}")
    
plt.xlabel("Time (turns)")
plt.ylabel("Std of gene_0")
plt.title("Std of gene_0 for different death factors")
plt.legend()
plt.show()
```

```python
# for genes:
plt.figure(figsize=(12, 8))
for i in range(6):
    plt.plot(data[i].time, data[i].gene_0_mean, label=f"Death factor = {0.5*i}")
    
plt.xlabel("Time (turns)")
plt.ylabel("Mean of gene_0")
plt.title("Mean of gene_0 for different death factors")
plt.legend()
plt.show()

# What happened to the purple graph - sudden increase in the mean of gene_0 -> investigate further? just a statistical anomaly?
# When the population dies down, it's random which genes survive, although very weird that this peak happened.
```

```python
raw_data2 = pd.read_csv('data/gene_game/oscillating_death_ratio_0_2.csv')
raw_data2
```

```python
data2 = raw_data2.copy()
```

```python
# for genes:

#define colors to use
col1 = 'black'
col2 = 'steelblue'

#define subplots
fig,ax = plt.subplots(figsize=(12, 8))

#add first line to plot
ax.plot(data2.time, data2.death_factor_1, label="Death factor", color=col1)
ax.set_xlabel("Time (turns)")
ax.set_ylim(0, 10)

#add y-axis label
ax.set_ylabel('Death factor', color=col1, fontsize=16)

#define second y-axis that shares x-axis with current plot
ax2 = ax.twinx()

ax2.plot(data2.time, data2.animal_number, label="Animal number", color=col2)


#add second y-axis label
ax2.set_ylabel('Animal number', color=col2, fontsize=16)


plt.title("Population number with oscillating death factor")
plt.legend()
plt.show()

```

```python
population_change = np.gradient(data2.animal_number, data2.time)
```

```python
# for genes:

#define colors to use
col1 = 'black'
col2 = 'steelblue'

#define subplots
fig,ax = plt.subplots(figsize=(12, 8))

#add first line to plot
ax.plot(data2.time, data2.death_factor_1, label="Death factor", color=col1)
ax.set_xlabel("Time (turns)")
ax.set_ylim(0, 10)

#add y-axis label
ax.set_ylabel('Death factor', color=col1, fontsize=16)
ax.legend()

#define second y-axis that shares x-axis with current plot
ax2 = ax.twinx()

ax2.plot(data2.time, data2.animal_number, color=col2, label="Total animal number")
ax2.plot(data2.time, population_change, color=col2, label="Gradient of animal number")



#add second y-axis label
ax2.set_ylabel('Animal number', color=col2, fontsize=16)


plt.title("Population number with oscillating death factor")
ax2.legend()
plt.show()

```
