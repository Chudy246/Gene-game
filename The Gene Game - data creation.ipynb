```python
# world with fully developed population experiencing sudden increase in a new disease - near extinction event
max_pop = 30000
animal_alive = np.zeros(max_pop)
animal_alive[:25000] = 1

animal_age = np.random.uniform(0, 40, size=max_pop)
animal_age = animal_age*animal_alive

animal_genes = np.random.uniform(0, 1, size=(max_pop, 2))
animal_genes = (animal_genes.T*animal_alive).T

animal_alive = pd.Series(animal_alive).apply(bool)
animal_age = pd.Series(animal_age).apply(int)
animal_genes = pd.DataFrame(animal_genes)
```

```python
# create columns for individual stats,
# aggregate into a list data from 200 turns
kill_factors = [-50, 2]
world = World(animal_alive, animal_age, animal_genes, kill_factors, printout=True)

data_columns = world.statistics(return_index=True)
data_raw = []
for j in range(7):
    print(f"Killing factor is equal to {0+0.5*j}")
    kill_factors = [-50, 0+0.5*j]  # uniform distribution from 0 to 3
    world = World(animal_alive, animal_age, animal_genes, kill_factors, printout=True)
    data_raw.append(world.statistics(return_index=False))
    for i in range(200):
        world.progress_time()
        data_raw.append(world.statistics(return_index=False))
```

```python
# create a dataset of population statistics
dataset = pd.DataFrame(data=data_raw, columns=data_columns)
dataset
```

```python
dataset.to_csv("data/gene_game/progressive_death_ratio_0_2,5.csv", index=False)
```

```python
# world with fully developed population experiencing periodic diseases - say winter season?
max_pop = 30000
animal_alive = np.zeros(max_pop)
animal_alive[:25000] = 1

animal_age = np.random.uniform(0, 40, size=max_pop)
animal_age = animal_age*animal_alive

animal_genes = np.random.uniform(0, 1, size=(max_pop, 2))
animal_genes = (animal_genes.T*animal_alive).T

animal_alive = pd.Series(animal_alive).apply(bool)
animal_age = pd.Series(animal_age).apply(int)
animal_genes = pd.DataFrame(animal_genes)
```

```python
x = np.linspace(0, 25*2*np.pi, 501)
death_factor = 1 - np.cos(x)
kill_factors = [-50, death_factor[0]]
world = World(animal_alive, animal_age, animal_genes, kill_factors, printout=True)
```

```python
data_columns = world.statistics(return_index=True)
data_raw = []

data_raw.append(world.statistics(return_index=False))
for i in range(500):
    world.progress_time()
    world.kill_factors = [-50, death_factor[i+1]]
    data_raw.append(world.statistics(return_index=False))
```

```python
# create a dataset of population statistics
dataset = pd.DataFrame(data=data_raw, columns=data_columns)
dataset
```

```python
dataset.to_csv("data/gene_game/oscillating_death_ratio_0_2.csv", index=False)
```
